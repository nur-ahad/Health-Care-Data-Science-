```python
#Import Libraries 
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
aapl = yf.Ticker("AAPL")
```

```python
#Getting historial stock price data
aapl_data = aapl.history(period="max")
```

```python
#Plotting the data
plt.plot(aapl_data["Close"])
plt.show()
```

```python
#50-day moving average 
aapl_data["50ma"] = aapl_data["Close"].rolling(window=50).mean()
```

```python
#Plot the closing prices and the 50-day moving average 
plt.plot(aapl_data["Close"],label ='Closing Price')
plt.plot(aapl_data["50ma"],label ='50-da moving average')
plt.legend()
plt.show()
```
